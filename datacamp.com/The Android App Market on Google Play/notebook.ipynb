```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
```

```python
apps=pd.read_csv("datasets/apps.csv")
print(apps.head())
print("....................................................................................")
print(apps.info())
print("....................................................................................")
print(apps.isna().sum())
#object datatype in pandas is similar to string data type
```

```python
#instruction..1 solution
apps['Installs'] = apps['Installs'].str.replace(',','')
apps['Installs'] = apps['Installs'].str.replace('+','')
apps['Installs'] = apps['Installs'].astype(int)
```

```python
#checking if instruction..1
print(apps.head())
print(apps.info())
```

```python
#instruction..2
#Save your answer as a DataFrame
app_category_info=apps.groupby("Category").agg({"App":"count","Price":"mean","Rating":"mean"})

#Your should rename the four columns as: Category, Number of apps, Average price, Average rating
app_category_info.rename(columns={'App':'Number of apps', 'Price':'Average price', 'Rating': 'Average rating'}, inplace=True)
```

```python
#checking if instruction..2
print(app_category_info.head())
```

```python
#instruction...3  .....................................................................................
#data overview
appReviews = pd.read_csv('datasets/user_reviews.csv')
appReviews.info()
appReviews.head()
```

```python
FINANCEapp=apps[apps['Category']=='FINANCE']
FINANCEapp=FINANCEapp[FINANCEapp["Type"]=="Free"]
print(FINANCEapp.head())
```

```python
#merge two dataFrame together
mergedAppDataframe=pd.merge(appReviews,FINANCEapp,on="App")
mergedAppDataframe
```

```python
#Your answer should have exactly 10 rows and two columns named: App and Sentiment Score, 
sentimentalScoredData=mergedAppDataframe.groupby("App")["Sentiment Score"].agg({'Sentiment Score':'mean'})

#where the average Sentiment Score is sorted from highest to lowest.
sentimentalScoredData=sentimentalScoredData.sort_values("Sentiment Score",ascending=False)

#answer should have exactly 10 rows
top_10_user_feedback=sentimentalScoredData.iloc[:10]
top_10_user_feedback
```
